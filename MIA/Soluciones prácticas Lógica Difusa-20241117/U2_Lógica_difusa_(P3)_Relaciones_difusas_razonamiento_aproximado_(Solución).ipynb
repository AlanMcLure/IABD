## 5) Relaciones difusas
Ejemplo de aplicación del <b>producto cartesiano</b> usando la librería "fuzzy.py"

```python
#
# Función cartesian(): Relación cartesiana difusa.
# Argumentos:
#   mA: numpy.ndarray, Vector con valores de pertenencia de xA
#   mB: numpy.ndarray, Vector con valores de pertenencia de xB
# Retorna:
#   AxB: Matriz de relación (mínimo)
#
x = np.array([0, 0.1, 0.5, 0.8, 1, 1])
y = np.array([0.8, 0.8, 0.9, 1])
cartesian(x, y)
```

#### Ejemplo: relación entre 2 conjuntos difusos mediante el producto cartesiano
A = Venta (moderada) de libros / B = Diversidad de la oferta

```python
# Representación del Conjunto A (Venta de libros) como función "campana de Bell generalizada"
xA = np.linspace(0, 100, 50)
mA = gbellmf(xA, [20, 3, 50])
plt.subplot(121)
plt.plot(xA, mA)

# Representación del Conjunto B (Diversidad de la oferta) como función sigmoide
xB = np.linspace(0, 1000, 50)
mB = sigmf(xB, [0.01, 500])
plt.subplot(122)
plt.plot(xB, mB)

# Representación en 3D del producto cartesiano
R_AB = cartesian(mA, mB)
A, B = np.meshgrid(xA, xB, indexing='ij')

fig = plt.figure()
ax = Axes3D(fig)
ax = plt.axes(projection ='3d')
ax.plot_surface(A, B, R_AB, cmap=plt.cm.hot)
ax.set_zlim(0, 1)
```

Ejemplo de aplicación de la <b>composición</b> usando la librería "fuzzy.py"

```python
#
# Función compose(): Composición difusa max-min.
# Argumentos:
#   mRA: numpy.ndarray, vector o matriz de relación A
#   mRB: numpy.ndarray, Matriz de relación B
# Retorna:
#   AoB: Matriz de composición max-min
#
x = np.array([[0.4, 1.0, 0.2],
              [0.1, 0.0, 0.5],
              [0.9, 0.7, 0.8]])
y = np.array([[1.0, 0.6, 0.5, 0.4, 0.0],
              [0.4, 0.8, 0.7, 0.3, 0.3],
              [0.5, 1.0, 0.5, 0.2, 0.8]])
compose(x, y)
```

#### Ejemplo: operación de composición entre 3 conjuntos difusos

```python
# Universo de discurso para x e y
x = np.linspace(0, 1, 50)
y = np.linspace(0, 5, 50)

# Funciones de membresía A(x) y B(y)
A = sigmf(x, [15, 0.5])
plt.subplot(221)
plt.title('A(x)')
plt.ylim((0, 1))
plt.plot(x, A)

B = sigmf(y, [-3, 2.5])
plt.subplot(222)
plt.title('B(y)')
plt.ylim((0, 1))
plt.plot(y, B)

# Producto cartesiano: AxB
R_AB = cartesian(A, B)
A, B = np.meshgrid(x, y, indexing='ij')


# Función de membresía Ac(x)
Ap = gaussmf(x, [0.04, 0.5])
plt.subplot(223)
plt.title('Ap(x)')
plt.ylim((0, 1))
plt.plot(x, Ap)

# Composición: Ac * (AxB)
Bp = compose(Ap, R_AB)
plt.subplot(224)
plt.title('Ap o (AxB)')
plt.ylim((0, 1))
plt.plot(y, Bp)

plt.subplots_adjust(wspace=0.5, hspace=0.5)

# Producto cartesiano en 3D
fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(A, B, R_AB, cmap=plt.cm.hot)
ax.set_zlim(0, 1)
```

```python
#
# Función Tmin(): Norma T
# Argumentos:
#   mA: numpy.ndarray, Vector con valores de pertenencia de xA
#   mB: numpy.ndarray, Vector con valores de pertenencia de xB
# Retorna:
#   Tmin: Norma T (mínimo)
#
x = np.linspace(0, 10, 50)
mA = gbellmf(x, [2, 3, 5])
plt.plot(x, mA)

mB = sigmf(x, [1, 5])
plt.plot(x, mB)

T = Tmin(mA, mB)
plt.plot(x, T, linewidth=3.0)
```

```python
#
# Función Smax(): Norma S
# Argumentos:
#   mA: numpy.ndarray, Vector con valores de pertenencia de xA
#   mB: numpy.ndarray, Vector con valores de pertenencia de xB
# Retorna:
#   Tmin: Norma S (máximo)
#
x = np.linspace(0, 10, 50)
mA = gbellmf(x, [2, 3, 5])
plt.plot(x, mA)

mB = sigmf(x, [1, 5])
plt.plot(x, mB)

T = Smax(mA, mB)
plt.plot(x, T, linewidth=3.0)
```

```python
#
# Función Nc(): Complemento
# Argumentos:
#   mA: numpy.ndarray, Vector con valores de pertenencia de xA
# Retorna:
#   Nc: Complemento
#
x = np.linspace(0, 10, 50)
A = gbellmf(x, [2, 3, 5])
plt.plot(x, A)

Ap = Nc(A)
plt.plot(x, Ap, linewidth=3.0)
```

<li><h4><b>Ejemplo</b>: Regla "Si estudio mucho, entonces soy excelente estudiante"</h4></li>

```python
# Se definen las variables:
x = np.linspace(0,8,50)   # Universo de discurso para horas diarias de estudio
y = np.linspace(1,10,50) # Universo de discurso para promedio de notas
mA = sigmf(x,[3,2]) # Función de membresía "estudio mucho"
mB = sigmf(y,[3,7.5]) # Función de membresía "excelente estudiante"
```

```python
# Graficamos para conocer la funciones
plt.subplot(1,2,1)
plt.title("estudio mucho")
plt.xlabel("horas")
plt.plot(x, mA)
plt.subplot(1,2,2)
plt.title("excelente estudiante")
plt.xlabel("promedio")
plt.plot(y,mB)
plt.show()
```

```python
# La relación A x B se define por la función cartesian()
R = cartesian(mA,mB)

# graficamos como contorno
A, B = np.meshgrid(x, y, indexing='ij')
plt.contourf(A,B,R, cmap=plt.cm.plasma)
plt.xlabel('horas de estudio')
plt.ylabel('calificación')
plt.show()
```

<h4>3. Modus Ponens difuso</h4>

```python
# Se definen los variables:
x = np.linspace(0,8,50) # horas diarias de estudio
y = np.linspace(1,10,50) # rango de calificaciones
mA = sigmf(x,[3,2]) # Función de membresía "estudia mucho"
mB = sigmf(y,[3,5.5]) # Función de membresía "excelente estudiante"

# Premisa 1: Hecho -> "Pedro estudia poco (Complemento de "estudia mucho")
Ap = Nc(mA) #

plt.subplot(221)
plt.plot(x,Ap)
plt.title("A' = Comp. A")

# Premisa 2: Regla -> "Si estudio mucho, entonces soy excelente estudiante"
R = cartesian(mA,mB)

plt.subplot(222)
A, B = np.meshgrid(x, y, indexing='ij')
plt.contourf(A,B,R, cmap=plt.cm.plasma)
plt.title('R = A x B')

# Conclusión: Bp = compose(Ap, R)
# Inferencia: Pedro tiene una baja pertenencia en el grupo de estudiantes excelentes
Bp = compose(Ap, R)

plt.subplot(212)
plt.plot(y, Bp)
plt.ylim(0,1)
plt.title("B' = A' o R")
plt.subplots_adjust(hspace=0.5)
plt.show()
```

```python
# Se definen las variables VA (velocidad creciente) y GA (gravedad del accidente)
VA = np.linspace(0,140,100) # velocidad creciente
GA = np.linspace(1,4,100) # gravedad del accidente

mVA = sigmf(VA,[0.1, 80]) # función de membresía para x
mGA = sigmf(GA, [2, 2.5]) # función de membresía para y

plt.figure(figsize = [10, 15])
plt.subplot(331)
plt.title("velocidad creciente")
plt.xlabel("valor")
plt.plot(VA, mVA)
plt.subplot(332)
plt.title("gravedad del accidente")
plt.xlabel("grado")
plt.plot(GA,mGA)

# Premisa 1 // Hecho: "Pedro no conduce a alta velocidad"
NmVA= Nc(mVA)

plt.subplot(334)
plt.title("baja velocidad")
plt.xlabel("valor")
plt.plot(VA,NmVA)

# Premisa 2 // Regla: "Si conduce a alta velocidad, el accidente será grave"
R = cartesian(mVA,mGA)

plt.subplot(335)
A, B = np.meshgrid(VA, GA, indexing='ij')
plt.contourf(A,B,R, cmap=plt.cm.plasma)
plt.title('R = mAV x mGA')

# Conclusión: Bp = compose(NmVA, R)
# Inferencia: Pedro tendrá un accidente lo menos grave posible
Bp = compose(NmVA, R)

plt.subplot(336)
plt.plot(GA, Bp)
plt.ylim(0,1)
plt.title("Conclusión")

plt.subplots_adjust(hspace=0.4, wspace=0.5)
plt.show()
```
